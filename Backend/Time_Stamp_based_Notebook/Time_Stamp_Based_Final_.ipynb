```python
!pip install moviepy

import re
from moviepy.editor import VideoFileClip

# Fixed paths
video_file = "/content/input/test.mp4"
time_file  = "/content/input/time.txt"
output_file = "/content/timestamp_requested.mp4"

# === Step 1: Read timestamps from time.txt ===
with open(time_file, "r") as f:
    line = f.read().strip()

match = re.match(r"(\d+:\d+)\s*to\s*(\d+:\d+)", line)
if not match:
    raise ValueError("❌ time.txt format must be 'MM:SS to MM:SS' (e.g., 4:00 to 8:00)")

start_time, end_time = match.groups()

# Convert MM:SS -> seconds
def to_seconds(t):
    parts = t.split(":")
    if len(parts) == 2:  # MM:SS
        return int(parts[0]) * 60 + int(parts[1])
    elif len(parts) == 3:  # HH:MM:SS (optional support)
        return int(parts[0]) * 3600 + int(parts[1]) * 60 + int(parts[2])
    else:
        raise ValueError("Invalid time format")

start_sec = to_seconds(start_time)
end_sec = to_seconds(end_time)

# === Step 2: Cut video using MoviePy ===
clip = VideoFileClip(video_file).subclip(start_sec, end_sec)
clip.write_videofile(output_file, codec="libx264", audio_codec="aac")

print(f"✅ Video trimmed from {start_time} to {end_time}")
print(f"👉 Saved as: {output_file}")
```
